```python
# Lets practice some linear regression
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats 
```

```python
x_values = np.random.normal(3, 1, 1000)
# print (x_values)
y_values = 100 - (x_values + np.random.normal(0, 1, 1000))*3

plt.scatter(x_values, y_values)
```

```python
slope, intercept, r_value, p_value, stderr = stats.linregress(x_values, y_values)
```

```python
# correlation of the best fit line
# if it close to 1 then good fir and close to 0 then it is a bad fit.
r_value ** 2
```

```python
# lets plot the points and line to check the best fit

def y_value(x):
    return intercept + slope*x
```

```python
plt.plot (x_values, y_value(x_values), c='r')
plt.scatter(x_values, y_values, c='b')
plt.show()
```
